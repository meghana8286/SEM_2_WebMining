networkx: A library for the creation, manipulation, and study of the structure, dynamics, and functions of complex networks.
matplotlib.pyplot: A plotting library for creating visualizations.

```python
import networkx as nx
import matplotlib.pyplot as plt
```

Create an empty directed graph using NetworkX.

```python
D = nx.DiGraph()
```

Add weighted edges to the graph. Each tuple represents an edge ('A' to 'B', 'A' to 'C', 'C' to 'A', 'B' to 'C') with a weight of 1.

```python
D.add_weighted_edges_from([('A', 'B', 1), ('A', 'C', 1), ('C', 'A', 1), ('B', 'C', 1)])
```

Print the PageRank of each node in the graph. PageRank is an algorithm used to measure the importance of nodes in a graph.

```python
print(nx.pagerank(D))
```

```python
# Specify the layout using a spring layout for better visualization
pos = nx.spring_layout(D)

# Draw the graph using NetworkX and Matplotlib
nx.draw(D, pos=pos, with_labels=True)

# Display the graph using Matplotlib
plt.show()
```
